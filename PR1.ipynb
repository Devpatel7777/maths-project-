```python
sns.histplot(data=df,x="Income", kde=True)
plt.show()
```

```python
from scipy.stats import norm

mean = df["Income"].mean()
std = df["Income"].std()
x = np.linspace(df["Income"].min(), df["Income"].max(), 100)
y = norm.pdf(x, mean, std)
sns.histplot(data = df, x = "Income" , kde=False,stat="density")
plt.plot(x, y)
plt.show()
```

```python
# Skewness 
df["Income"].skew()
```

```python
# Kurtosis
df["Income"].kurt()
```

Boxplot

```python
sns.boxplot(x="Education", y="Income", data=df)
plt.title("Income by Education Level")
plt.show()
```

```python
plt.figure()
sns.boxenplot(data = df , x = "Area", y = "Income")
plt.title("Income by Area")
plt.show()
```
